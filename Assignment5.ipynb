# **Q1. Create a simple page with some div tags and show different ways to add CSS as well as what happens when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the project where required.**

**The three types of CSS:**
1.   Inline CSS
2.   Internal CSS
3.   External CSS



**What happens when we target the same elements with Inline, Internal and External CSS?**
1.   Inline CSS overrides both internal and external CSS.
2.   Internal CSS overrides external CSS but is overridden by inline.
3.   External CSS has the lowest priority.
4.   All styles still merge, unless they override the same property.


**code11.html**
```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Adding CSS to our web-page</title>
    <!-- 2. Internal CSS -->
    <style>
      #inter {
        color: orange;
        text-transform: uppercase;
        text-decoration: underline red wavy 1px;
        line-height: 50px;
      }

      #all{
        color: red;
      }
    </style>
    <!-- 3. External CSS -->
    <link rel="stylesheet" href="code11.css" />
  </head>
  <body>
    <!-- 1. Inline CSS -->
    <div style="color: brown; font-size: xx-large">Pulkit Dubey</div>

    <!-- 2. Internal CSS using id-->
    <div id="inter">Pulkit Dubey</div>

    <!-- 3. External CSS -->
    <div id="exter">Pulkit Dubey</div>

    <!-- targeting the same elements with inline, internal, and external CSS -->
    <div
      id="all"
      style="font-size: 40px; color: black; text-shadow: 2px 2px 2px orange; line-height: 150px;"
    >
      Bholenath Computer Academy
    </div>
  </body>
</html>

```

**code11.css**


```
#exter{
    color: yellow;
    font-size: 40px;
    font-weight: bold;
    text-shadow: 2px 2px 2px black;
}
#all{
    color: blue;
}
```
