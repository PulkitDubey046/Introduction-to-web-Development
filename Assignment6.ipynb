```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <title>Google Fonts Blog</title>

    <!-- Google Fonts -->
    <link
      href="https://fonts.googleapis.com/css2?family=Montserrat&family=Oswald&display=swap"
      rel="stylesheet"
    />

    <style>
      body {
        margin: 0;
        padding: 0;
        font-family: "Montserrat", sans-serif;
        background-color: #f5f5f5;
      }

      h1 {
        font-family: "Oswald", sans-serif;
        text-align: center;
        margin-top: 40px;
        font-size: 48px;
        color:  #333 ;
      }

      .content {
        max-width: 800px;
        margin: 30px auto;
        padding: 20px;
        background-color: #ffffff;
        box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1);
        text-align: center;
      }

      .content img {
        width: 100%;
        max-width: 600px;
        height: auto;
        margin-bottom: 20px;
        border-radius: 8px;
      }

      .content p {
        font-size: 18px;
        line-height: 1.6;
        color: #444;
        font-family: "Montserrat", sans-serif;
      }
    </style>
  </head>
  <body>
    <h1>Google Fonts</h1>

    <div class="content">
      <img
        src="https://news.sophos.com/wp-content/uploads/2022/01/gf-1200.png?w=640"
        alt="Google Fonts Logo"
      />

      <p>
        Google Fonts is a free and open-source web font service provided by
        Google. It allows designers and developers to use a wide variety of
        fonts in their websites and applications with minimal effort. The
        service is fast, reliable, and easy to integrate via a simple `&lt; link
        / &gt` in HTML. With a vast collection of typefaces like Montserrat,
        Oswald, Roboto, and more, Google Fonts helps bring beautiful typography
        to the web without compromising performance.
      </p>
    </div>
  </body>
</html>

```
