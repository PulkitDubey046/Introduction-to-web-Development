## Purpose of var() in CSS
The var() function in CSS is used to insert the value of a custom property (CSS variable). It promotes reusability, maintainability, and consistency in styling. Instead of hardcoding values like colors, fonts, or spacings multiple times, you define them once as variables and reuse them with var().



```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Document</title>
    <style>
      :root {
        --primaryColor: #00b7ff;
        --secondaryColor: #6c757d;
        --textColor:white;
      }

      #primaryBtn {
        background-color: var(--primaryColor);
        color: var(--textColor);
        padding: 10px 20px;
        border: none;
      }

      #secondaryBtn {
        background-color: var(--secondaryColor);
        color:var(--textColor);
        padding: 10px 20px;
        border: none;
      }
    </style>
  </head>
  <body>
    <button id="primaryBtn">Primary Button</button>
    <button id="secondaryBtn">Secondary Button</button>
  </body>
</html>

```

**Problem Statement**

Create a 3D cube using the transform property of CSS.

image.png



```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Document</title>
    <style>
      body {
        display: flex;
        justify-content: center;
        align-items: center;
        height: 100vh;
        margin: 0;
        background: #f0f0f0;
      }

      .scene {
        width: 200px;
        height: 200px;
        perspective: 600px;
      }

      .cube {
        width: 100%;
        height: 100%;
        position: relative;
        transform-style: preserve-3d;
        transform: rotateX(20deg) rotateY(30deg);
      }

      .face {
        position: absolute;
        width: 200px;
        height: 200px;
        background: rgba(0, 183, 255, 0.8);
        border: 1px solid #000;
        display: flex;
        align-items: center;
        justify-content: center;
        font-weight: bold;
      }

      .front {
        transform: translateZ(100px);
      }
      .back {
        transform: rotateY(180deg) translateZ(100px);
      }
      .right {
        transform: rotateY(90deg) translateZ(100px);
      }
      .left {
        transform: rotateY(-90deg) translateZ(100px);
      }
      .top {
        transform: rotateX(90deg) translateZ(100px);
      }
      .bottom {
        transform: rotateX(-90deg) translateZ(100px);
      }
    </style>
  </head>
  <body>
    <div class="scene">
      <div class="cube">
        <div class="face front">Front</div>
        <div class="face back">Back</div>
        <div class="face right">Right</div>
        <div class="face left">Left</div>
        <div class="face top">Top</div>
        <div class="face bottom">Bottom</div>
      </div>
    </div>
  </body>
</html>

```

**Problem Statement**

Create a simple circular loader which will rotate continuously to look like a loading screen on a website.

image.png




```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <style>
         body{
            display: flex;
            flex-wrap: wrap;
            justify-content: center;
            align-items: center;
            height: 100vh;
            margin: 0;
            background-color: aliceblue;
        }
        .loader{
            width: 50px;
            height: 50px;
            margin: 20px;
        }
        #loader1{       
            border: 8px solid #f3f3f3;
            border-top: 8px solid #3498db;
            border-radius: 50%;
            animation: spin 2s linear infinite;
        }
        @keyframes spin {
            0%{
                transform: rotate(0deg);
            }
            100%{
                transform: rotate(360deg);
            }
        }
    </style>
</head>
<body>
    <div class="loader" id="loader1"></div>
</body>
</html>
```
