**Problem Statement**

# Explain the use of grid-auto-row and grid-auto-column using code examples.

In CSS Grid, we usually define how many rows and columns we want using grid-template-rows and grid-template-columns.

But sometimes, the content overflows  meaning more rows or columns are created automatically because we didn't define enough.

That's where:

**grid-auto-rows**  sets the height of those extra rows

**grid-auto-columns**  sets the width of those extra columns

# **grid-auto-rows**

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Document</title>
    <style>
      .grid {
        display: grid;
        grid-template-columns: 150px 150px;
        grid-auto-rows: 100px;
        gap: 10px;
      }
      .box {
        background-color: lightgreen;
        font-size: 18px;
        text-align: center;
        line-height: 100px;
      }
    </style>
  </head>
  <body>
    <div class="grid">
      <div class="box">1</div>
      <div class="box">2</div>
      <div class="box">3</div>
      <div class="box">4</div>
    </div>
  </body>
</html>
```



# **grid-auto-columns**

**Problem Statement**

# Explain the difference between justify-items and justify-self using code examples.

In CSS Grid, justify-items and justify-self are used to control the horizontal alignment of items inside grid cells. The **justify-items** property is applied to the grid container and affects all the items inside it. For example, if you use `justify-items: center`, all the grid items will be centered horizontally inside their individual cells.

On the other hand, **justify-self** is used on individual grid items to control their alignment separately. This property overrides justify-items for that specific item. So if all items are aligned to the left using justify-items: start, you can still move one item to the right using `justify-self: end`. This gives you more control when you want different items to be aligned differently inside the same grid.


# **justify-items**
```
<!DOCTYPE html>
<html>
<head>
  <style>
    .grid {
      display: grid;
      grid-template-columns: repeat(2, 100px);
      justify-items: center; /* All items centered */
      gap: 10px;
      border: 2px solid black;
      padding: 10px;
    }

    .box {
      background-color: lightgreen;
      padding: 10px;
      border: 1px solid green;
      text-align: center;
    }
  </style>
</head>
<body>

  <div class="grid">
    <div class="box">A</div>
    <div class="box">B</div>
  </div>

</body>
</html>

```

# **justify-self**



```
<!DOCTYPE html>
<html>
<head>
  <style>
    .grid {
      display: grid;
      grid-template-columns: repeat(2, 100px);
      justify-items: start; /* Default: items aligned to the left */
      gap: 10px;
      border: 2px solid black;
      padding: 10px;
    }

    .box {
      background-color: lightblue;
      padding: 10px;
      border: 1px solid blue;
      text-align: center;
    }

    .special {
      justify-self: end; /* This one item moves to the right */
    }
  </style>
</head>
<body>

  <div class="grid">
    <div class="box">A</div>
    <div class="box special">B</div>
  </div>

</body>
</html>

```
