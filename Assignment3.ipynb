```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <title>HTML Tags List</title>
  </head>
  <body>
    <h1>Ordered List of HTML Tags</h1>
    <ol>
      <li>
        <strong>&lt;html&gt;</strong>  This tag wraps all the content on the
        entire web page as the root element.
      </li>
      <li>
        <strong>&lt;head&gt;</strong>  This tag holds meta-information about
        the document such as the title and links to CSS or JS files.
      </li>
      <li>
        <strong>&lt;title&gt;</strong>  This tag sets the web page title shown
        in the browser tab.
      </li>
      <li>
        <strong>&lt;body&gt;</strong>  This tag holds all the visible content
        of the web page.
      </li>
      <li>
        <strong>&lt;h1&gt; to &lt;h6&gt;</strong>  These header tags create
        headings, with &lt;h1&gt; being the largest and &lt;h6&gt; the smallest.
      </li>
      <li><strong>&lt;p&gt;</strong>  This tag creates a text paragraph.</li>
      <li>
        <strong>&lt;a&gt;</strong>  This tag makes a hyperlink to another page
        or website.
      </li>
      <li>
        <strong>&lt;img&gt;</strong>  This tag puts images into the web page.
      </li>
      <li>
        <strong>&lt;ul&gt;</strong>  This tag makes an unordered (bulleted)
        list.
      </li>
      <li>
        <strong>&lt;ol&gt;</strong>  This tag makes an ordered (numbered) list.
      </li>
      <li>
        <strong>&lt;li&gt;</strong>  This tag represents a list item in ordered
        or unordered lists.
      </li>
      <li>
        <strong>&lt;div&gt;</strong>  This block-level container groups
        elements for styling or layout.
      </li>
      <li>
        <strong>&lt;span&gt;</strong>  This inline container styles a part of
        text or groups inline elements.
      </li>
    </ol>
  </body>
</html>

```

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <title>Full Stack Web Development Tech Stack</title>
  </head>
  <body>
    <h1>Full Stack Web Development Tech Stack</h1>

    <dl>
      <dt><strong>Fundamentals</strong></dt>
      <dd>
        Basic understanding of computers, internet, programming logic, and how
        the web works.
      </dd>

      <dt><strong>HTML</strong></dt>
      <dd>
        The backbone of all websites, used to create and structure content on
        the web.
      </dd>

      <dt><strong>CSS</strong></dt>
      <dd>
        Used to design and style HTML elements, controlling layout, colors,
        fonts, and more.
      </dd>

      <dt><strong>Tailwind CSS</strong></dt>
      <dd>
        A utility-first CSS framework that makes styling faster using
        pre-defined classes.
      </dd>

      <dt><strong>JavaScript</strong></dt>
      <dd>
        Programming language that brings interactivity to websites (e.g.,
        sliders, forms, dynamic content).
      </dd>

      <dt><strong>React JS</strong></dt>
      <dd>
        A JavaScript library for building user interfaces using a
        component-based structure.
      </dd>

      <dt><strong>Redux</strong></dt>
      <dd>
        A state management tool for React applications, useful for handling
        complex data flows.
      </dd>

      <dt><strong>Git</strong></dt>
      <dd>
        A version control system that tracks code changes and helps developers
        collaborate.
      </dd>

      <dt><strong>GitHub</strong></dt>
      <dd>
        An online platform to host code using Git, share projects, and
        collaborate with others.
      </dd>

      <dt><strong>Node.js</strong></dt>
      <dd>
        Allows JavaScript to be used on the server-side to build scalable
        backend services.
      </dd>

      <dt><strong>Express JS</strong></dt>
      <dd>
        A fast and minimal web framework for Node.js used to build APIs and
        backend applications.
      </dd>

      <dt><strong>DBMS</strong></dt>
      <dd>
        Database Management System  software used to create, manage, and
        interact with databases.
      </dd>

      <dt><strong>MongoDB</strong></dt>
      <dd>
        A NoSQL database that stores data in flexible, JSON-like documents.
      </dd>

      <dt><strong>Postman</strong></dt>
      <dd>
        A tool for testing APIs by sending requests and viewing responses
        easily.
      </dd>

      <dt><strong>MongoDB Compass</strong></dt>
      <dd>
        A graphical interface for MongoDB that allows users to visualize and
        manage their data.
      </dd>

      <dt><strong>Vercel</strong></dt>
      <dd>
        A platform to quickly deploy and host frontend applications, especially
        React-based projects.
      </dd>

      <dt><strong>Netlify</strong></dt>
      <dd>
        An easy-to-use platform for deploying and hosting static websites and
        frontend apps.
      </dd>

      <dt><strong>Render</strong></dt>
      <dd>
        A cloud platform that lets you deploy both frontend and backend apps
        with databases.
      </dd>
    </dl>
  </body>
</html>

```
