# **TASK 1**

**Problem Statement**

**Q1. Explain the potential SEO challenges and solutions associated with Single Page Applications (SPAs).**

**SEO Challenges in SPAs**

- Content is loaded with JavaScript, making it hard for crawlers to index.
- SPAs often lack unique URLs for different pages.
- Dynamic meta tags may not be visible to search engines.
- Initial page load can be slow, hurting SEO.
- Social sharing previews may not display correctly.


**SEO Solutions for SPAs**

- Use Server-Side Rendering (SSR) to serve full HTML to crawlers.
- Apply pre-rendering tools like Prerender.io.
- Ensure unique URLs with proper client-side routing.
- Manage meta tags with tools like React Helmet.
- Improve performance using lazy loading and code-splitting.

# **TASK 2**

**Problem Statement**

Explain the key differences and use cases between React's Client-Side Rendering (CSR) and Server-Side Rendering (SSR). Provide examples of scenarios where each approach is advantageous, and discuss the challenges associated with using React in both contexts.




**Client-Side Rendering (CSR)**

- Rendering happens in the browser after JavaScript loads.
- Good for dynamic, interactive apps (e.g., dashboards, social media).
- Faster navigation after initial load since only data updates.
- SEO can be weak as crawlers may miss JS-rendered content.
- **Example:** Single Page Applications like Facebook or Trello.

**Server-Side Rendering (SSR)**

- Pages are rendered on the server and sent as HTML to the browser.
- Great for SEO-focused sites (e.g., blogs, e-commerce).
- Faster first page load and better crawlability.
- More complex setup and higher server load.
- **Example:** Next.js apps for news websites or product listings.

# **TASK 3**

**Problem Statement**

**Explain the difference between npm and npx using the create-react-app code example.**

**Difference between npm and npx :**

**NPM (Node Package Manager)** is used to install packages globally or locally in your project. For example, if you want to use create-react-app via npm, you first install it globally:
```
npm install -g create-react-app
create-react-app my-app
```
**npx** is a package runner that lets you execute packages without installing them globally. Using npx with create-react-app runs the tool directly:

```
npx create-react-app my-app
```
This downloads and runs the latest version of create-react-app without needing a global install.

So, npm manages package installations, while npx helps you run packages easily without global installs, making it ideal for tools like create-react-app.

# **TASK 4**

**Problem Statement**

Create a React component called "Login" without using JSX for its definition. This component should render an HTML structure that includes a form with input fields for a username and password. It should also have a button labeled "Submit."

image.png

```
import React from "react";

function Login() {
  return React.createElement(
    "form",
    null,
    [
      React.createElement("input", {
        type: "text",
        placeholder: "username",
        key: "username"
      }),
      React.createElement("input", {
        type: "password",
        placeholder: "password",
        key: "password"
      }),
      React.createElement(
        "button",
        { type: "submit", key: "submit" },
        "Submit"
      )
    ]
  );
}

export default Login;
```
