# **Q1. Describe the main differences between the CSS Flexbox layout model and the CSS Grid layout model. When would you choose to use one over the other?**

**Answer :**

####  Main Differences between the CSS Flexbox layout model and the CSS Grid layout model

| Feature              | Flexbox                                      | Grid                                              |
|----------------------|----------------------------------------------|---------------------------------------------------|
| Layout Direction     | One-dimensional (row **or** column)          | Two-dimensional (row **and** column)              |
| Content Flow         | Follows content size                         | Based on grid tracks (defined by developer)       |
| Alignment            | Great for aligning items along one axis      | Precise alignment in rows and columns             |
| Use Case             | Smaller UI components (e.g. navbars)         | Entire page layouts or complex grids              |

---

#### When to Use Flexbox
- One-dimensional layouts
- Content flows in one direction
- Need simple alignment/distribution
- **Example**: nav menus, toolbars, card layouts

---

#### When to Use Grid
- Two-dimensional layouts (rows + columns)
- Layouts require structure and alignment on both axes
- **Example**: dashboards, full page layouts, image galleries

---



# **Q2. Explain the role of the following key properties in the Flexbox layout model**
1. justify-content
2. align-item
3. gap
4. flex-direction
5. flex-wrap

**Answer :**


#### 1. `justify-content`

Moves the items left, right, or center in a row.

**Examples**:
- `flex-start`: all items go to the left
- `flex-end`: all items go to the right
- `center`: items go to the middle
- `space-between`: space between items only
- `space-around`: space around all items
- `space-evenly`: equal space between and around all items

---

#### 2. `align-items`

Moves the items up and down in the container.

**Examples**:
- `stretch`: items grow to fill the height
- `flex-start`: items go to the top
- `flex-end`: items go to the bottom
- `center`: items go to the middle
- `baseline`: items line up by the text line

---

#### 3. `gap`

Adds space between items.

**Example**:
```css
gap: 10px;
```
This adds 10 pixels of space between each item.

---

#### 4. `flex-direction`

Decides the direction items go.

**Examples**:
- `row`: left to right (default)
- `row-reverse`: right to left
- `column`: top to bottom
- `column-reverse`: bottom to top

---

#### 5. `flex-wrap`

Tells items if they should move to the next line if there's no space.

**Examples**:
- `nowrap`: stay on one line
- `wrap`: move to next line when needed
- `wrap-reverse`: move to next line in reverse order

---
