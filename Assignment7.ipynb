#**Task 1:**


**Problem Statement**

Write code to position 5 equal-sized (50 X 50px) boxes A, B, C, D, and E as follows.
  1. box A 200px from the left, and 200px from the top of the viewport.
  2. box B -30px left, and -30px above from the center of box A.
  3. box B -30px right, and -30px above from the center of box A.
  4. box B -30px left, and -30px below from the center of box A.
  5. box B -30px right, and -30px below from the center of box A.



# **Task 2:**


**Problem Statement**

Explain the difference between Absolute and Relative positioning.


# **Answer:**

**Relative positioning** means the element stays in its place but moves a little from there, like moving up, down, left, or right. It still takes up space in the page.

**Absolute positioning** means the element is taken out from its place and put exactly where you want it on the page, based on its parent box. It does not take up any space where it was before.

So, relative is like moving a toy on a table a bit, and absolute is like lifting the toy and placing it somewhere else.



# **Task 5:**


**Problem Statement**

Explain the z-index, with a code example.

# Answer

The **z-index** property in CSS controls the stacking order of elements along the z-axis (which is front-to-back in 2D space). Elements with a higher z-index value appear in front of those with a lower value, but only if the element has a position value other than static (e.g., relative, absolute, or fixed).

Imagine you are drawing on transparent sheets and placing them one on top of another. The sheet on the top will cover the ones below it, right? Thats exactly what z-index does in websites.

When we have things (like boxes or images) on a web page that can overlap, z-index tells which one should be in front.




```
<!DOCTYPE html>
<html>
<head>
  <style>
    .box {
      width: 150px;
      height: 150px;
      position: absolute;
    }

    .box1 {
      background-color: red;
      top: 50px;
      left: 50px;
      z-index: 1;
    }

    .box2 {
      background-color: blue;
      top: 100px;
      left: 100px;
      z-index: 2;
    }
  </style>
</head>
<body>

  <div class="box box1">Box 1</div>
  <div class="box box2">Box 2</div>

</body>
</html>

```
